```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#fit studies the corpus and keeps an array of unique words, give index to it
#vect is being trained on this corpus
vect = CountVectorizer(binary=True)
sent = ["HELLOOOOOO HI I WANT TO SLEEP PLEASE","I REALLY REALLY NEED SLEEP"]
vect.fit(sent)
```

```python
#vocab is a dictionary with it's own index value and words
vocab = vect.vocabulary_
for key in sorted(vocab.keys()):
    print("{}:{}".format(key,vocab[key]))
```

```python
print(vect.transform(["Ugh i need sleep"]).toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vect.transform(["Helloo i am a sentence"]).toarray(), vect.transform(["Greetings, This is a sentence"]).toarray() )
print(similarity)
```

```python
def gender_feature(word):
    return{'last_letter':word[-1]}
```

```python
from nltk.corpus import names
names.words()
```

```python
print(len(names.words()))78
```

```python
labeled_name=[(name,'male') for name in names.words('male.txt')] +[(name,'female') for name in names.words('female.txt')]
```

```python
labeled_name
```

```python
import random
random.shuffle(labeled_name)
```

```python
featuresets =[(gender_feature(n),gender) for (n,gender) in labeled_name]
featuresets
```

```python
train_set,test_set = featuresets[:5000],featuresets[5000:]
```

```python
import nltk
classifier= nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_feature('Parvathy'))
print(nltk.classify.accuracy(classifier,test_set))
```
